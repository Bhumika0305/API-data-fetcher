```python
import requests

API_URL = "https://jsonplaceholder.typicode.com/posts"


def fetch_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        print("Successfully fetched data!")
        return response.json()

    except requests.exceptions.HTTPError as http_err:
        print(f"HTTP error: {http_err} (Status Code: {response.status_code})")
    except requests.exceptions.ConnectionError:
        print("Connection error: Unable to reach API")
    except requests.exceptions.Timeout:
        print("Timeout error: Request took too long")
    except requests.exceptions.RequestException as err:
        print(f"Unexpected error: {err}")
    except ValueError:
        print("Error decoding JSON response")

    return []


def display_posts(posts, limit=3):
    if not posts:
        print("No posts to display.")
        return

    print(f"\nShowing {min(limit, len(posts))} posts:\n")

    for i, post in enumerate(posts[:limit]):
        print(f"Post {i+1}:")
        print(f"  ID: {post['id']}")
        print(f"  User ID: {post['userId']}")
        print(f"  Title: {post['title'][:50]}...")
        print(f"  Body: {post['body'][:100]}...")
        print("-" * 30)


def filter_by_user(posts, user_id):
    return [post for post in posts if post['userId'] == user_id]


def search_posts(posts, keyword):
    return [post for post in posts if keyword.lower() in post['title'].lower()]


def main():
    posts = fetch_data(API_URL)

    if not posts:
        return

    print(f"\nTotal posts fetched: {len(posts)}")

    print("\nChoose an option:")
    print("1. Show first 3 posts")
    print("2. Filter by User ID")
    print("3. Search by keyword")

    choice = input("Enter choice (1/2/3): ")

    if choice == '1':
        display_posts(posts)

    elif choice == '2':
        user_id = int(input("Enter User ID: "))
        filtered = filter_by_user(posts, user_id)

        print(f"\nPosts by User ID {user_id}: {len(filtered)} found")
        display_posts(filtered)

    elif choice == '3':
        keyword = input("Enter keyword: ")
        results = search_posts(posts, keyword)

        print(f"\nSearch results for '{keyword}': {len(results)} found")
        display_posts(results)

    else:
        print("Invalid choice")


if __name__ == "__main__":
    main()
```
